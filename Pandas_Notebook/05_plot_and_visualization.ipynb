```python
# Sample DataFrame
np.random.seed(42)
df = pd.DataFrame({
    'Apples': np.random.randint(20, 50, 7),
    'Bananas': np.random.randint(30, 60, 7),
    'Cherries': np.random.randint(15, 35, 7)
}, index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
df
```

##  1. Line Plot (default)

```python
df.plot(title = "Line plot of Fruits")
plt.ylabel("Amount sold")
plt.xlabel("Days of a week")
plt.show()
```

## 2. Bar plot (vertical)

```python
df.plot(kind = "bar", title = "Bar plot of Fruits")
plt.ylabel("Amount sold")
plt.xlabel("Days of a week")
plt.show()
```

## 3. Bar plot (horizontal)

```python
df.plot(kind = "barh", title = "Bar plot of Fruits")
plt.xlabel("Amount sold")
plt.ylabel("Days of a week")
plt.show()
```

##  4. Histogram (on 1 column)

```python
df['Apples'].plot(kind = "hist", bins = 12)
plt.show()
```

##  5. KDE Plot

```python
df["Cherries"].plot(kind = 'kde')
plt.show()
```

## 6. Box plot

```python
df.plot(kind = "box")
plt.show()
```

## 7. Area plot

```python
df.plot(kind = 'area')
plt.show()
```

## 8. Scatter plot

```python
df.plot(kind = 'scatter', x = 'Apples', y = 'Bananas')
plt.show()
```

# Customizing plots

## 1. Figure size and title

```python
df.plot(figsize = (8, 6), title = "Fruit sales by day")
plt.ylabel("Amount sold")
plt.xlabel("Days")
plt.show()
```

## 2. Colors and line styles

```python
df.plot(color = ['red', 'yellow', 'green'], linestyle = '--', marker = 'x')
plt.show()
```

## 3. Axis labels and legend

```python
ax = df.plot()
ax.set_xlabel("Days")
ax.set_ylabel("Amount sold")
ax.legend(title = "Fruits")
plt.show()
```

## 4. Gridlines and Axis limits

```python
df.plot(grid = True, ylim = (0, 100))
plt.show()
```

# Subplots and Multiple Axes

```python
df.plot(subplots = True, layout = (3, 1), figsize = (10, 8),
       title = "Week sales by day", sharey = True)
plt.tight_layout()
plt.show()
```

```python
np.random.seed(123)
fruit = pd.DataFrame({
    'Apples': np.random.randint(20, 60, 7),
    'Oranges': np.random.randint(25, 65, 7),
    'Grapes': np.random.randint(15, 40, 7)
}, index=['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
```

```python
fruit
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (12, 8))


# Top-left: Line Plot
fruit.plot(kind='line', ax=axes[0, 0], title='Line Plot of Fruit Sales', marker='o', color=['green', 'orange', 'purple'])
axes[0, 0].set_ylabel('Sales')

# Top-right: Bar Plot
fruit.plot(kind='bar', ax=axes[0, 1], title='Bar Plot of Fruit Sales', color=['red', 'yellow', 'blue'])
axes[0, 1].set_ylabel('Quantity Sold')

# Bottom-left: Box Plot
fruit.plot(kind='box', ax=axes[1, 0], title='Box Plot of Fruit Sales', color='cyan')
axes[1, 0].set_ylabel('Sales')

# Bottom-right: Scatter Plot
fruit.plot(kind='scatter', x='Apples', y='Oranges', ax=axes[1, 1], title='Apples vs Oranges', color='brown')
axes[1, 1].set_xlabel('Apples')
axes[1, 1].set_ylabel('Oranges')

# Adjust layout so the subplots don't overlap
plt.tight_layout()
plt.show()
```
